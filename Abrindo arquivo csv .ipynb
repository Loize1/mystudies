Abrindo arquivo CSV 

```python
a = open("Salary_dataset.csv", "r") #abrindo arquivo em modo leitura
```

```python
dados = a.read() #gravando na variável dados
```

```python
rows = dados.split('\n') #separando as linhas a partir do enter e transformando tudo em uma linha só
#rows é do tipo lista
```

```python
print(rows)
```

```python
full_data = [] #criação de uma lista vazia
```

```python
for row in rows:
    split_row = row.split(",")
    full_data.append(split_row)
#Para cada linha (row) na lista de linhas (rows), separe as linhas por vírgula
#O resultado será adicionado à lista, até então vazia, full_data
#Criação de uma lista de listas. Cada linha corresponde a uma lista
print(full_data)
```

```python
count = 0
for row in full_data:
    count += 1 #Mesma coisa que count = count + 1
#Criação de um contador para saber  quantidade de linhas
```

```python
print(count)
```
